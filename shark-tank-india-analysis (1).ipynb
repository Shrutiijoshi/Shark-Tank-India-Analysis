<div style="background-color: #4682B4; padding: 20px; border-radius: 10px;">
      <h1 style="color: #fff;"><strong>Trends in deal-making over different episodes</strong></h1>
</div>

```python
deals_per_episode=shark_tank_df.groupby("episode_number")["deal"].sum()
```

```python
deals_per_episode
```

```python
plt.plot(deals_per_episode.index,deals_per_episode.values,color="#08306b")
plt.title('Number of Deals Made per Episode')
plt.xlabel('Episode Number')
plt.ylabel('Number of Deals')
plt.show()
```

<div style="background-color: #4682B4; padding: 20px; border-radius: 10px;">
      <h1 style="color: #fff;"><strong>Ideas of businesses attract the most investment</strong></h1>
</div>

```python
category_investment=shark_tank_df[shark_tank_df["deal"]==1].groupby("idea")["deal_amount"].sum().sort_values(ascending=False).head(10)
```

```python
category_investment
```

```python
category_investment.plot(kind="bar",color="#08306b")
plt.title("Total Investment by Business Ideas")
plt.xlabel("Business Idea")
plt.ylabel("Total Investment Amount")
plt.xticks(rotation=45,ha="right")
plt.show()
```

<div style="background-color: #4682B4; padding: 20px; border-radius: 10px;">
      <h1 style="color: #fff;"><strong>Relationship between the Amount asked and the equity offered</strong></h1>
</div>

```python
sns.scatterplot(data=shark_tank_df, x='pitcher_ask_amount', y='ask_equity',color="#08306b")
plt.title('Relationship between Amount Asked and Equity Offered')
plt.xlabel('Amount Asked (in Lakhs)')
plt.ylabel('Equity Offered (%)')
plt.show()
```

<div style="background-color: #4682B4; padding: 20px; border-radius: 10px;">
      <h1 style="color: #fff;"><strong>Multi-Shark Deal</strong></h1>
</div>

```python
multi_sharks_df=shark_tank_df[shark_tank_df["total_sharks_invested"]>1].sort_values("total_sharks_invested",ascending=False).head(10)
```

```python
sns.barplot(data=multi_sharks_df,x="total_sharks_invested",y="brand_name",palette="Blues_r")
plt.xlabel('Number of Sharks Invested')
plt.ylabel('Brand Name')
plt.title('Brands with Multiple Sharks Invested in Shark Tank India')
plt.show()
```

<div style="background-color: #4682B4; padding: 20px; border-radius: 10px;">
      <h1 style="color: #fff;"><strong>Total Investment in 5-Sharks Deals</strong></h1>
</div>

```python
plt.figure(figsize=(7,5))
total_sharks=shark_tank_df[shark_tank_df["total_sharks_invested"]==5]
sns.barplot(total_sharks,x="brand_name",y="deal_amount",color="#08306b")
plt.title("5-Shark deal and Total investment")
plt.xlabel("Brand Name")
plt.ylabel("Deal Amount")
plt.show()
```

<div style="background-color: #4682B4; padding: 20px; border-radius: 10px;">
      <h1 style="color: #fff;"><strong>Number of Deals done by Individual Sharks</strong></h1>
</div>

```python
num_shark_deal = [shark_tank_df.ashneer_deal.sum(), shark_tank_df.anupam_deal.sum(), shark_tank_df.aman_deal.sum(), 
                  shark_tank_df.namita_deal.sum(), shark_tank_df.vineeta_deal.sum(), shark_tank_df.peyush_deal.sum(), 
                  shark_tank_df.ghazal_deal.sum()]

all_sharks = ["Ashneer", "Anupam", "Aman", "Namita", "Vineeta", "Peyush", "Ghazal"]

colors = ['#6baed6', '#2171b5', '#08306b', '#4292c6', '#9ecae1', '#08519c', '#c6dbef']
explode = (0, 0, 0.1, 0, 0, 0, 0)  
plt.figure(figsize=(8, 8)) 
plt.pie(num_shark_deal, labels=all_sharks, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Number of Deals by Each Shark")
plt.show()
```
